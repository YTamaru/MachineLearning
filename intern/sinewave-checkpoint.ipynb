```python
def get_n_sequence_dataset_and_m_peripd_later_data(data, n_prev = 100, m = 1):  
   """
   data should be pd.DataFrame()
   """

   docX, docY = [], []
   for i in range(len(data)-n_prev*m):

       docX.append(data.iloc[i:i+n_prev].as_matrix())
       docY.append(data.iloc[i+n_prev+(m-1)].as_matrix())

   alsX = np.array(docX)
   alsY = np.array(docY)

   return alsX, alsY
```

```python
def create_train_data_and_test_data(df, test_size=0.1, n_prev = 100, m = 1):  
   """
   This just splits data to training and testing parts
   """
   ntrn = round(len(df) * (1 - test_size))
   ntrn = int(ntrn)
   X_train, y_train = get_n_sequence_dataset_and_m_peripd_later_data(df.iloc[0:ntrn], n_prev, m)
   X_test, y_test = get_n_sequence_dataset_and_m_peripd_later_data(df.iloc[ntrn:], n_prev, m)

   return (X_train, y_train), (X_test, y_test)
```

```python
import tensorflow
import pandas as pd
import numpy as np
import math
import random
import seaborn as sns
```

```python
# 乱数の種として、0を渡す
random.seed(0)
random_factor = 0.05
# sin曲線 1周期あたりのステップ数
steps_per_cycle = 100
# 生成する周期数
number_of_cycles = 1000

# 区間 -1.0 〜 +1.0 の 一様乱数 付き の sin曲線 
df = pd.DataFrame(np.arange(steps_per_cycle * number_of_cycles + 1), columns=["t"])
df["sin_t"] = df.t.apply(lambda x: math.sin(x * (2 * math.pi / steps_per_cycle)
                                               + random.uniform(-1.0, +1.0) * random_factor))
```

```python
# 2周期分、描画
df[["sin_t"]].head(steps_per_cycle * 2).plot()
```

```python
# 1周期分、描画
df[["sin_t"]].head(steps_per_cycle).plot()
```

```python
# 1周期＝100時点
print(steps_per_cycle)
```

```python
# 1時点先の値を予測する回帰モデルの学習用及び検証用データを作成
# 直近 1.5周期分の値推移パターン を手掛かり　に、その1時点先 の（１点の）値 を 予測する回帰モデル
(X_train, y_train), (X_test, y_test) = create_train_data_and_test_data(df[["sin_t"]], 
                                                                        n_prev=int(steps_per_cycle*1.5),
                                                                        m=1)  
```

```python
# 生成されたデータセットの件数
print("学習用 説明変数データセット数: ", len(X_train))
print("学習用 目的変数データセット数: ", len(y_train))
print("============================================")
print("検証用 説明変数データセット数: ", len(X_test))
print("検証用 目的変数データセット数: ", len(y_test))
```

```python
# 各データセットの中身の要素数
print("学習用 説明変数データ要素数: ", len(X_train[0]))
print("学習用 目的変数データ要素数: ", len(y_train[0]))
print("============================================")
print("検証用 説明変数データ要素数: ", len(X_test[0]))
print("検証用 目的変数データ要素数: ", len(y_test[0]))
```

```python
from keras.models import Sequential  
from keras.layers.core import Dense, Activation  
from keras.layers.recurrent import LSTM

# 中間層（LSTM層）のLSTMブロックの数は、300にする
#
# batch_input_shape=(None, int(steps_per_cycle)*1.5, 1) の 意味は、以下になる。
#
# batch_input_shape=(None, input_length=int(steps_per_cycle)*1.5, input_dim=1)
# batch_input_shape=(None,\
#                   入力データの要素（ノード）数（LSTM層のLSTMブロックの数に一致する）（※文書データなら単語の数。今回は説明変数の数）,\
#                   LSTM中間層に投入するデータの次元数（※文書データなら１次元配列なので1)
#                  )
# （参考） http://qiita.com/HirofumiYashima/items/a4651229865918ae2a78

model_predict_1_ahead = Sequential()  
model_predict_1_ahead.add(LSTM(300, batch_input_shape=(None, int(steps_per_cycle)*1.5, 1), return_sequences=False))  
model_predict_1_ahead.add(Dense(1))  
model_predict_1_ahead.add(Activation("linear"))  
model_predict_1_ahead.compile(loss="mean_squared_error", optimizer="rmsprop")
```

```python
# 入力データのサイズ は、steps_per_cycle * 1 にする
# 【 説明変数 】X_train[0:steps_per_cycle]
# 【 目的変数 】y_train[0:steps_per_cycle]
#
# これは、モデル に 与える学習データ として、
# 「目的変数：１時点のデータ」及び「その直前の1.5周期分のデータパターン」
# を、与える際、
# 目的変数（１時点のデータ）が、sin曲線１周期 内の任意の点である場合
# の（それに対応する）（説明変数）「その直前の1.5周期分のデータパターン」を、
# モデルが学習する機会を与えるため。

from keras.callbacks import EarlyStopping
early_stopping = EarlyStopping(monitor='val_loss', patience=0)

model_1_ahead =   model_predict_1_ahead.fit(X_train[0:steps_per_cycle], y_train[0:steps_per_cycle], 
                                                    batch_size=int(len(X_train)/6), 
                                                    nb_epoch=10, 
                                                    validation_split=0.05,
                                                    callbacks=[early_stopping])
```

```python
loss = model_1_ahead.history['loss']
val_loss = model_1_ahead.history['val_loss']

nb_epoch = len(loss)

plt.plot(range(nb_epoch), loss, marker='.', label='loss')
plt.plot(range(nb_epoch), val_loss, marker='.', label='val_loss')

plt.legend(loc='best', fontsize=10)
plt.grid()

plt.xlabel('epoch')
plt.ylabel('loss')
plt.show()
```
